```python
    def longestPalindrome(self, s):
        n = len(s)
        nn = 2 * n + 1
        sn = ["\0"] * nn
        j = 0
        for i in range(0, nn):
            if i % 2 == 1:
                sn[i] = s[j]
                j += 1
        L = [0] * nn
        pd = 0
        mx = 0
        ml = 0
        mp = 0
        for i in range(1, nn):
            if i >= mx:
                pd = i
                j = i - 1
                k = i + 1
                while j >= 0 and k < nn:
                    if sn[j] != sn[k]:
                        break
                    j -= 1
                    k += 1
                mx = k - 1
                L[i] = mx - pd
                if L[i] > ml:
                    ml = L[i]
                    mp = i
                continue
            ibar = 2 * pd - i
            if L[ibar] < mx - i:
                L[i] = L[ibar]
            elif L[ibar] > mx - i:
                L[i] = mx - i
            else:
                pd = i
                j = i - 1
                k = i + 1
                while j >= 0 and k < nn:
                    if sn[j] != sn[k]:
                        break
                    j -= 1
                    k += 1
                mx = k - 1
                L[i] = mx - pd
                if L[i] > ml:
                    ml = L[i]
                    mp = i
        ts = ""
        for i in range(mp + 1, mp + L[mp] + 1):
            if sn[i] != "\0":
                ts += sn[i]
        ts1 = list(ts)
        ts1.reverse()
        tsf = "".join(ts1)
        if sn[mp] != "\0":
            re = tsf + sn[mp] + ts
        else:
            re = tsf + ts
        return re
```
